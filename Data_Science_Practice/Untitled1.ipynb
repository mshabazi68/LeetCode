```python
## read all the file in the directory and then we will concat (merge in to one file)
files = [file for file in os.listdir('./Sales_Data')]
all_months_data = pd.DataFrame()

for file in files:
    df= pd.read_csv ("./Sales_Data/" + file)
    all_months_data = pd.concat([all_months_data , df])

all_months_data.to_csv("all_data.csv" , index = False)

```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

####  Agument the Data with additional columns 

#### Task2:  Add Month Column


```python
## just adding the new column called month in to our data
all_data['Month'] = 3
all_data.head()
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data.head()
```


### Deleting The Nan value and cleaning

```python
## here we finding all the NAN values in the file ! 
nan_df = all_data [all_data.isna().any(axis=1)]
nan_df.head()
```

## Dropping the NaN Value form the table 

```python
# Dropping all the NAN values that we found erliear in the table 
all_data = all_data.dropna(how='all')
all_data.head()
```

## Fixing the ERROR of the table when we have 'Or'

```python
## we find 'Or' and DELETE it right away
all_data = all_data[all_data['Order Date'].str[0:2] != "Or"]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Convert Columns to the correct type 

```python
# here we try to conver the following data to the (INT or FLOAT) data type because this 
all_data['Quantity Ordered'] = pd.to_numeric(all_data ['Quantity Ordered']) # Make int 
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # Make float 
```

#### Add a sales column 

```python
all_data['sales'] = all_data['Quantity Ordered'] * all_data ['Price Each']
all_data.head()
```
